```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

chelsea = pd.read_excel('Chelsea.xlsx')
man = pd.read_excel('United.xlsx')
city = pd.read_excel('City.xlsx')
# city
```

```python
def clean(x):
    return x[0:4]
yearc=list(chelsea['Season'])
yearc = [x for x in yearc if str(x) != 'nan']
resultc=list(map(clean,yearc))

resultc = [i for i in resultc if int(i)>1905]
# print(resultc)
# resultc[80:101]
```

```python
yearm=list(man['Season']+" ")
yearm = [x for x in yearm if str(x) != 'nan']
resultm=list(map(clean,yearm))

resultm = [i for i in resultm if int(i)>1905]
# resultm[81:102]
```

```python
yearcity=list(city['Season']+" ")
yearcity = [x for x in yearcity if str(x) != 'nan']
resultcity=list(map(clean,yearcity))

resultcity = [i for i in resultcity if int(i)>1905]
# resultcity[90:110]
# print(resultm)
```
